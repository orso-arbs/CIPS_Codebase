```python
parts = {
    'part_name': ['part1', 'part2', 'part3', 'part4', 'part5', 'part6'],
    'state_names_in_part': [[list(range(1, 50+1))], [list(range(1, 11+1))], [list(range(1, 15+1))], [list(range(1, 11+1))], [list(range(1, 26+1))], [list(range(1, 23+1))]],

    'state uncompressed size measured [GB]': np.array([56255808136, 56255808136, 109662528136, 109662528136, 106007232136, 98285442184]) / 1024**3,

    }
parts_df = pd.DataFrame(parts)

# Display the entire updated DataFrame


# Flatten the list inside each list (there's an extra nesting level)
parts_df['num_states'] = parts_df['state_names_in_part'].apply(lambda x: len(x[0]))
parts_df['total_size'] = parts_df['num_states'] * parts_df['state uncompressed size measured [GB]']
M_total_gb = parts_df['total_size'].sum()

display(parts_df)
M_scratch_max = 2.5 * 1024
print(f"Total uncompressed size of all states: {M_total_gb/1024:.2f} TB")
N_scrathces_min = M_total_gb / M_scratch_max
print(f"N_scrathces_min = {N_scrathces_min}")


######################################################################## States
print("\nStates calc:")

States = {
    'part': [],
    'state_names_in_part': [],
    'state_name_in_total': [],
    'state uncompressed size measured [GB]': [],
}
States_df = pd.DataFrame(States)

# Initialize empty list to build States_df rows
states_rows = []
# Global counter for state_name_in_total
state_counter = 1

for idx, row in parts_df.iterrows():
    part_name = row['part_name']
    state_names = row['state_names_in_part'][0]  # Unwrap the nested list
    state_size_measured = row['state uncompressed size measured [GB]']
    #compressed_size = row['state compressed size measured [GB]']
    #uncompressed_estimate = row['state uncompressed size estimate [GB]']

    for state_name in state_names:
        state_row = {
            'part': part_name,
            'state_names_in_part': state_name,
            'state_name_in_total': state_counter,
            'state uncompressed size measured [GB]': state_size_measured
        }
        states_rows.append(state_row)
        state_counter += 1

# Create DataFrame
States_df = pd.DataFrame(states_rows)

# Display result
#display(States_df)
```

```python
import pandas as pd

# --- Initialize scratch_df ---
scratch_df = pd.DataFrame([
    { # david
        "location": "/cluster/scratch/doancea/orsoMT_doancea",
        "states in scratch": [],
        "#states": 0.0,
        "total memory available [GB]": 2.5 * 1024 - 4.1 / 1024**2,
        "total memory used by states [GB]": 0.0,
        "percent free when used": 0.0,
    },
    { # leon
        "location": "/cluster/scratch/scleon/orsoMT_scleon",
        "states in scratch": [],
        "#states": 0.0,
        "total memory available [GB]": 2.5 * 1024 - 4.1 / 1024**2,
        "total memory used by states [GB]": 0.0,
        "percent free when used": 0.0,
    },
    { # alex
        "location": "/cluster/scratch/aolareanu/orsoMT_aolareanu",
        "states in scratch": [],
        "#states": 0.0,
        "total memory available [GB]": 2.5 * 1024 - 598.02 / 1024**2,
        "total memory used by states [GB]": 0.0,
        "percent free when used": 0.0,
    },
    { # lorenz
        "location": "/cluster/scratch/loherzog/orsoMT_loherzog",
        "states in scratch": [],
        "#states": 0.0,
        "total memory available [GB]": 2.5 * 1024 - 8.19 / 1024**2,
        "total memory used by states [GB]": 0.0,
        "percent free when used": 0.0,
    },
    { # seba
        "location": "/cluster/scratch/nobss01/orsoMT_nobss01",
        "states in scratch": [],
        "#states": 0.0,
        "total memory available [GB]": 2.5 * 1024 - 720.9 / 1024**2,
        "total memory used by states [GB]": 0.0,
        "percent free when used": 0.0,
    },
    { # orso
        "location": "/cluster/scratch/orsob/orsoMT_orsob",
        "states in scratch": [],
        "#states": 0.0,
        "total memory available [GB]": 2.5 * 1024,
        "total memory used by states [GB]": 0.0,
        "percent free when used": 0.0,
    },
])

N_scratch = len(scratch_df["location"])
print(f"N_scratch = {N_scratch}")

scratch_df['states by part'] = [{} for _ in range(len(scratch_df))]

# --- First-fit allocation under 95% cap ---
memory_loading_percent = 0.95
for idx, state in States_df.iterrows():
    if pd.notnull(state['state uncompressed size measured [GB]']):
        size = state['state uncompressed size measured [GB]']
        for sidx, scratch in scratch_df.iterrows():
            used = scratch_df.at[sidx, 'total memory used by states [GB]']
            capacity = scratch_df.at[sidx, 'total memory available [GB]'] * memory_loading_percent

            if used + size <= capacity:
                # Assign location
                scratch_location = scratch['location']
                States_df.at[idx, 'location'] = scratch_location
                scratch_df.at[sidx, 'states in scratch'].append(state['state_name_in_total'])

                # Temporarily store part and state name for later part_dir mapping
                part = state['part']
                orig_state = state['state_names_in_part']
                scratch_df.at[sidx, 'states by part'].setdefault(part, []).append(orig_state)

                # Update usage stats
                scratch_df.at[sidx, 'total memory used by states [GB]'] = used + size
                scratch_df.at[sidx, '#states'] = len(scratch_df.at[sidx, 'states in scratch'])
                scratch_df.at[sidx, 'percent free when used'] = (
                    (scratch_df.at[sidx, 'total memory available [GB]'] - scratch_df.at[sidx, 'total memory used by states [GB]'])
                    / scratch_df.at[sidx, 'total memory available [GB]'] * 100
                )
                break

# --- Build sub-directory names for each (part, storage_location) pair ---
part_locations = {}
for part in States_df['part'].unique():
    locs = States_df.loc[States_df['part'] == part, 'location'] \
                    .dropna().unique().tolist()
    part_locations[part] = {
        loc: f"{part}_{chr(ord('a') + i)}"
        for i, loc in enumerate(locs)
    }

# --- Add 'part_dir' and 'file_location' columns ---
def get_part_dir(row):
    if pd.isna(row['location']):
        return None
    return part_locations[row['part']][row['location']]

States_df['part_dir'] = States_df.apply(get_part_dir, axis=1)
States_df['file_location'] = States_df.apply(
    lambda row: row['location'].rstrip('/') + '/' + row['part_dir']
    if pd.notna(row['location']) and pd.notna(row['part_dir']) else None,
    axis=1
)

# --- Final update of 'states by part' using part_dir ---
# Reset and rebuild with proper subpart naming
scratch_df['states by part'] = [{} for _ in range(len(scratch_df))]
for idx, state in States_df.iterrows():
    part_dir = state['part_dir']
    orig_state = state['state_names_in_part']
    loc = state['location']

    sidx = scratch_df[scratch_df['location'] == loc].index[0]
    scratch_df.at[sidx, 'states by part'].setdefault(part_dir, []).append(orig_state)
```

```python
# Create a new scratch_df from the altered States_df
scratch_df_fromManuallyAltered = pd.DataFrame(columns=[
    "location",
    "states in scratch",
    "#states",
    "total memory used by states [GB]",
    "states by part",
    "total memory available [GB]",
    "percent free when used"
])

for location, group in States_df_manuallyAltered.groupby('location'):
    # Calculate the total memory used by states for the location
    total_memory_used = group["state uncompressed size measured [GB]"].sum()
    
    # Calculate the number of states
    num_states = len(group)
    
    # Group states by part directory
    states_by_part = group.groupby("part_dir")["state_names_in_part"].apply(list).to_dict()

    # Define the total available memory for the scratch directory (you can adjust this as needed)
    total_memory_available = 2.5 * 1024  # Assuming 2.5 TB or 2.5 * 1024 GB as example
    
    # Calculate the percentage of free memory
    percent_free_when_used = (total_memory_available - total_memory_used) / total_memory_available * 100
    
    # Append the data to the new DataFrame
    scratch_df_fromManuallyAltered = pd.concat([
        scratch_df_fromManuallyAltered,
        pd.DataFrame([{
            "location": location,
            "states in scratch": group["state_name_in_total"].tolist(),
            "#states": num_states,
            "total memory used by states [GB]": total_memory_used,
            "states by part": states_by_part,
            "total memory available [GB]": total_memory_available,
            "percent free when used": percent_free_when_used
        }])
    ], ignore_index=True)
```

```python
# Ensure that both dataframes have the same index order
scratch_df_fromManuallyAltered = scratch_df_fromManuallyAltered.set_index('location', drop=False)
scratch_df = scratch_df.set_index('location', drop=False)

# Reorder `scratch_df_fromManuallyAltered` based on the index order of `scratch_df`
scratch_df_fromManuallyAltered = scratch_df_fromManuallyAltered.reindex(scratch_df.index)

# Ensure the columns are ordered the same in both DataFrames
scratch_df_fromManuallyAltered = scratch_df_fromManuallyAltered[scratch_df.columns]


# Reset index to a numerical range starting from 0
scratch_df = scratch_df.reset_index(drop=True)
scratch_df_fromManuallyAltered = scratch_df_fromManuallyAltered.reset_index(drop=True)

# Display both DataFrames in the same order (rows and columns)
print("\n Computationally distributed")
display(scratch_df)

print("\n Manually Altered")
display(scratch_df_fromManuallyAltered)
```
